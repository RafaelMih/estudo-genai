Visualize embeddings, explore similaridade semântica e compare modelos.

```python
import sys; sys.path.insert(0, '..')
from modules.embeddings.local_embedder import LocalEmbedder
emb = LocalEmbedder()
print(f'Modelo: {emb.model_name}, Dimensão: {emb.dimension}')
```

## 1. Similaridade Semântica

```python
from modules.embeddings.similarity import cosine_similarity
pairs = [
    ('RAG usa busca vetorial', 'Retrieval-Augmented Generation combina busca com LLMs'),
    ('RAG usa busca vetorial', 'Python é uma linguagem de programação'),
    ('O gato dormiu', 'O felino repousou'),
]
for a, b in pairs:
    va, vb = emb.embed_one(a), emb.embed_one(b)
    sim = cosine_similarity(va, vb)
    print(f'{sim:.4f}  |  "{a[:40]}" ↔ "{b[:40]}"')
```

## 2. Busca Semântica

```python
from modules.embeddings.similarity import semantic_search
corpus = [
    'RAG combina recuperação com geração de texto',
    'ChromaDB é um banco vetorial open-source',
    'Python é amplamente usado em ML',
    'Transformers usam self-attention',
    'Embeddings representam semântica em vetores',
]
query = 'Como funciona a busca semântica?'
results = semantic_search(query, corpus, emb, top_k=3)
print(f'Query: "{query}"\n')
for r in results:
    print(f'  [{r["score"]:.4f}] {r["text"]}')
```
