# 01 — LLM Exploration

Experimentos com parâmetros de sampling, streaming e entrada multimodal.

```python
import sys; sys.path.insert(0, '..')
from shared.llm_client import build_client
client = build_client()
print('Cliente pronto!')
```

## 1. Efeito do Temperature

```python
prompt = 'Continue: A IA generativa é como...'
messages = [{'role': 'user', 'content': prompt}]
for temp in [0.0, 0.3, 0.7, 1.0]:
    resp = client.complete(messages, temperature=temp, max_tokens=80)
    print(f'temp={temp}: {resp.strip()[:120]}')
```

## 2. Streaming de Tokens

```python
import sys
messages = [{'role': 'user', 'content': 'Explique embeddings em 3 parágrafos.'}]
for token in client.stream(messages):
    print(token, end='', flush=True)
print()
```

## 3. Conversa Multi-turn

```python
history = []
def chat(message):
    history.append({'role': 'user', 'content': message})
    reply = client.complete(history, system='Você é um tutor de IA.')
    history.append({'role': 'assistant', 'content': reply})
    return reply

print(chat('O que é RAG?'))
print('---')
print(chat('Como ele se compara ao fine-tuning?'))
```
