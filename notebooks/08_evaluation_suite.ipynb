## 1. Métricas de String

```python
from modules.evaluation.metrics.string_metrics import rouge_score, bleu_score, f1_token_overlap
pred = 'RAG combina busca vetorial com geração de texto para melhorar respostas.'
ref = 'RAG usa recuperação de informações para aumentar a qualidade das respostas de LLMs.'
print('ROUGE:', rouge_score(pred, ref))
print('BLEU: ', bleu_score(pred, ref))
print('F1:   ', f1_token_overlap(pred, ref))
```

## 2. Similaridade Semântica

```python
from modules.evaluation.metrics.semantic_metrics import embedding_similarity
sim = embedding_similarity(pred, ref)
print(f'Similaridade semântica (embedding cosine): {sim}')
```

## 3. LLM-as-Judge

```python
import json
from modules.evaluation.metrics.llm_judge import LLMJudge
judge = LLMJudge()
scores = judge.evaluate(
    question='O que é RAG?',
    answer=pred,
    context='RAG (Retrieval-Augmented Generation) combina busca de documentos com LLMs.'
)
print(json.dumps(scores, ensure_ascii=False, indent=2))
```
